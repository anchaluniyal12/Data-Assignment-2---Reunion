Load the data and perform transformations to simplify the data and store it a normalized manner into smaller tables which are easier to analyze

```python
programs_data = []
concerts_data = []
works_data = []
soloists_data = []

# Iterate over the JSON data to extract information
for program in df['programs']:
    # Extract program details
    program_id = program['programID']
    season = program['season']
    orchestra = program['orchestra']
    for concert in program['concerts']:
        concerts_data.append({
            'programID': program_id,
            'Date': concert['Date'],
            'eventType': concert['eventType'],
            'Venue': concert['Venue'],
            'Location': concert['Location'],
            'Time': concert['Time']
        })
        for work in program['works']:
          work_data = {
            'ID': work['ID'],
            'workTitle': work.get('workTitle', None),
            'composerName': work.get('composerName', None),
            'conductorName': work.get('conductorName', None),
            'movement': work.get('movement', None)
        }
        works_data.append(work_data)

        for soloist in work.get('soloists', []):
            soloists_data.append({
                'workID': work['ID'],
                'soloistName': soloist['soloistName'],
                'soloistRoles': soloist['soloistRoles'],
                'soloistInstrument': soloist['soloistInstrument']
            })
```

```python
df_programs = pd.DataFrame(df['programs'])
df_concerts = pd.DataFrame(concerts_data)
df_works = pd.DataFrame(works_data)
df_soloists = pd.DataFrame(soloists_data)
```

```python
import sqlite3
conn = sqlite3.connect('orchestra_data.db')
cursor = conn.cursor()

# Create tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS Programss(
    programID TEXT PRIMARY KEY,
    season TEXT,
    orchestra TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Concerts (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    programID TEXT,
    Date TEXT,
    eventType TEXT,
    Venue TEXT,
    Location TEXT,
    Time TEXT,
    FOREIGN KEY (programID) REFERENCES Program (programID)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Works (
    ID TEXT PRIMARY KEY,
    workTitle TEXT,
    composerName TEXT,
    conductorName TEXT,
    movement TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Soloists (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    workID TEXT,
    soloistName TEXT,
    soloistRoles TEXT,
    soloistInstrument TEXT,
    FOREIGN KEY (workID) REFERENCES Work (ID)
)
''')
```

```python
cursor.execute('DROP TABLE IF EXISTS Works')
```

```python
# Insert data into tables


# Insert data into Concert table
for program in df['programs']:
    for concert in program['concerts']:
        cursor.execute('''
        INSERT INTO Concerts (programID, Date, eventType, Venue, Location, Time)
        VALUES (?, ?, ?, ?, ?, ?)
        ''', (program['programID'], concert['Date'], concert['eventType'], concert['Venue'], concert['Location'], concert['Time']))









# Insert data into Soloist table
for program in df['programs']:
    for work in program['works']:
        for soloist in work.get('soloists', []):
            cursor.execute('''
            INSERT INTO Soloists (workID, soloistName, soloistRoles, soloistInstrument)
            VALUES (?, ?, ?, ?)
            ''', (work['ID'], soloist.get('soloistName', None), soloist.get('soloistRoles', None), soloist.get('soloistInstrument', None)))

# Commit and close the connection
conn.commit()
conn.close()
```
